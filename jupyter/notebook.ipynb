```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('../filtered_data/acidentes_filtrado_full.csv', sep=';')
df.head()
```

```python
morte_ano = df.filter(items=['ano', 'mortos']).groupby('ano').sum().reset_index()

plt.bar(morte_ano['ano'], morte_ano['mortos'])
plt.title('Distribuição de mortes por ano')
plt.xlabel('ano')
plt.ylabel('mortos')
plt.show()
display(morte_ano)
```

```python
morte_mes = df.filter(items=['mes', 'mortos']).groupby('mes').sum().sort_values(by='mortos').reset_index()

fig, ax = plt.subplots()
ax.pie(morte_mes['mortos'], labels=morte_mes['mes'], autopct='%1.1f%%')
```
