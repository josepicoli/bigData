```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
tabela = pd.read_csv("../acidentes2024.csv", sep=";", encoding="Windows-1252")

estadoDesejado = "PI"
tabelaFiltrada = tabela[tabela["uf"] == estadoDesejado]
display(tabelaFiltrada)
```

```python
plt.hist(tabelaFiltrada["idade"], bins=20)
plt.title("Distribuição de idades em acidentes no PI")
plt.xlabel("idade")
plt.ylabel("Frequência")
plt.show()
```
